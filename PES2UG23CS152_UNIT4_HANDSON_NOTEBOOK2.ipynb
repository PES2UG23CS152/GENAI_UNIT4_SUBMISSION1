```python
import os

# 🔥 SET THESE
GROQ_API_KEY = ""
TAVILY_API_KEY = ""

# AutoGen needs this specifically
os.environ["OPENAI_API_KEY"] = GROQ_API_KEY
os.environ["TAVILY_API_KEY"] = TAVILY_API_KEY

print("✅ Keys set")
```

```python
from autogen import ConversableAgent
from tavily import TavilyClient

groq_config = [
    {
        "model": "llama-3.3-70b-versatile",
        "api_key": GROQ_API_KEY,
        "base_url": "https://api.groq.com/openai/v1",
        "api_type": "openai",
    }
]

llm_config = {
    "config_list": groq_config,
    "temperature": 0.3,
}

print("✅ LLM ready")
```

```python
def run_web_search(query: str, max_results: int = 5) -> str:
    client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))

    response = client.search(query=query, max_results=max_results)

    results = []
    for r in response.get("results", []):
        results.append(
            f"Title: {r.get('title', '')}\n"
            f"URL: {r.get('url', '')}\n"
            f"Content: {r.get('content', '')}"
        )

    return "\n\n---\n\n".join(results)

print("✅ Search ready")
```

```python
# User proxy (controller)
user_proxy = ConversableAgent(
    name="User",
    llm_config=False,
    human_input_mode="NEVER",
    code_execution_config=False,
    default_auto_reply="Continue or say TERMINATE."
)

# Researcher
researcher = ConversableAgent(
    name="Researcher",
    llm_config=llm_config,
    system_message=(
        "You are a research analyst. Given web search results, create a structured "
        "report with key facts, statistics, and sources. End with TERMINATE."
    )
)

# Writer
writer = ConversableAgent(
    name="Writer",
    llm_config=llm_config,
    system_message=(
        "You are a content writer. Convert research into an engaging article with "
        "sections, headings, and takeaways. End with TERMINATE."
    )
)

# Translator
translator = ConversableAgent(
    name="Translator",
    llm_config=llm_config,
    system_message=(
        "Translate the article into Spanish while preserving structure. End with TERMINATE."
    )
)

print("✅ Agents ready")
```

```python
TOPIC = "Recent advances in large language models in 2024 and 2025"

# Step 1: Search
print("🔎 Searching...")
search_results = run_web_search(TOPIC)

# Step 2: Research
print("\n🧠 Researching...")
research_chat = user_proxy.initiate_chat(
    researcher,
    message=f"Topic: {TOPIC}\n\n{search_results}",
    max_turns=3
)

research_output = research_chat.chat_history[-1]["content"].replace("TERMINATE", "").strip()

# Step 3: Writing
print("\n✍️ Writing article...")
writer_chat = user_proxy.initiate_chat(
    writer,
    message=research_output,
    max_turns=3
)

article = writer_chat.chat_history[-1]["content"].replace("TERMINATE", "").strip()

# Step 4: Translation
print("\n🌍 Translating...")
translator_chat = user_proxy.initiate_chat(
    translator,
    message=article,
    max_turns=3
)

translated = translator_chat.chat_history[-1]["content"].replace("TERMINATE", "").strip()

print("\n✅ Pipeline complete")
```

```python
print("\n" + "="*80)
print("📊 RESEARCH OUTPUT\n")
print(research_output)

print("\n" + "="*80)
print("📝 ARTICLE\n")
print(article)

print("\n" + "="*80)
print("🌍 TRANSLATION (SPANISH)\n")
print(translated)
```
